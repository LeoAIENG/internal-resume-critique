```python
from ruamel.yaml import YAML
from io import StringIO
from pathlib import Path

def yaml2dict(y):
	return YAML().load(y)

def yaml_read(path):
	_yaml = yaml2dict(Path(path).read_text())
	return _yaml

def dict2yaml(d):
	output_stream = StringIO()
	YAML().dump(d, output_stream)
	return output_stream.getvalue()

def yaml_dumps(d, path):
	with open(path, "w") as f:
		YAML().dump(d, f)
```

```python
def change_prompt(prompt, path):
	_yaml = yaml2dict(Path(path).read_text())
	_yaml["prompt"]["user"] = prompt
	up_yaml = dict2yaml(_yaml)
	Path(path).write_text(up_yaml)
```

```python
## MODEL
ORGANIZATION = "multimodalai"
CLIENT = "resume-critique"
MODEL = "llama3_1_8b"
MODEL_NUMBER = "1"
REV_N = "3"
FT_METHOD = "tt_lora"
MODEL_TYPE = "adapter"

MDATA_ID = f"model_{MODEL_NUMBER}_20k"
REV = f"rev_{REV_N}"
FT_MODEL_NAME = f"{CLIENT}-{MODEL}-{FT_METHOD}-{MDATA_ID}-{MODEL_TYPE}-{REV}"
FT_MODEL_HF_ID = f"{ORGANIZATION}/{FT_MODEL_NAME}"

## DATASET
TRAINING_DATA = "resume_critique_model_1.jsonl"

## PATH
MODEL_FOLDER = "model_1_rev_3/"
OUTPUT_MODEL_PATH = MODEL_FOLDER + f"checkpoint/{ORGANIZATION}/{FT_MODEL_NAME}"
TRAINING_DATA_PATH = f"data/{TRAINING_DATA}"
CONFIG_FILE_PATH = MODEL_FOLDER + f"config/{CONFIG_FILE}"
```

```python
model = AutoPeftModelForCausalLM.from_pretrained(OUTPUT_MODEL_PATH, torch_dtype=torch.float16, device_map="cuda:0")
```

```python
tokenizer = AutoTokenizer.from_pretrained(
	BASE_MODEL_PATH,
	padding_side = "right",
	add_eos_token = True
)
tokenizer.pad_token = tokenizer.eos_token
```

```python
gen_params = {
	"temperature": 0.45,
	"max_new_tokens": 300,
    "top_p": 0.90
}
```

```python
def generate_llm(prompt, gen_params):
	inputs = tokenizer(prompt, return_tensors = "pt").to("cuda:0")
	generate_ids = model.generate(inputs.input_ids, **gen_params)
	returned = tokenizer.batch_decode(generate_ids, skip_special_tokens = False)[0]
	generated_text = returned.split("### Response:\n")[1]
	generated_text = generated_text.rsplit(".", 1)[0] + "."
	return generated_text
```
