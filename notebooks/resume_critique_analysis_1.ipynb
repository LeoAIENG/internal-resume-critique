````python
# TEMPLATES
def get_templates():
	'''Get inputs for Main APP'''
	templates_d = yaml.load((DATA_EXTERNAL_PATH / "templates.yaml").read_text(), Loader=Loader)
	return templates_d

TEMPLATES = get_templates()["input"]
````

````python
def get_summary_input(positions, profession, cur_position):
	final_summary_input = TEMPLATES["summary"].format(
		  cur_pos=cur_position,
		  profession=profession,
		  positions=positions
	)
	return final_summary_input
````

````python
def get_positions(work_exps):
	positions = ""
	n_position = 0
	for work_exp in work_exps:
		title = work_exp["title"]
		employer = work_exp["employer"]
		description = work_exp["description"] if not "generated_jd" in work_exp else work_exp["generated_jd"]
		position = TEMPLATES["summary_skills_position"].format(
			 n_position=(n_position:=n_position+1),
			 title=title,
			 employer=employer,
			 description=description
		)
		positions += position
	return positions
````

````python
def format_raw_data_resume(file_n, raw_data_item):
	## NAME
	first_name = raw_data_item["first_name"]
	middle_name = raw_data_item["middle_name"]
	last_name = raw_data_item["last_name"]
	NAME = " ".join([first_name, middle_name, last_name])

	## SUMMARY
	SUMMARY = raw_data_item["summary"]
	
	## PROFESSION
	PROFESSION = raw_data_item["profession"]
	
	## POSITIONS
	positions_output_l = raw_data_item["positions"]
	
	# Loop Outputs
	work_exps = []
	for out_idx, position_output in enumerate(positions_output_l):
		output_t_nrm = position_output["title_normalized"]
		output_t = position_output["title"]
		output_org = position_output["organization"]
		output_desc = position_output["description"]

		# Final Org, Title, Description and Response
		final_org = output_org
		final_title = output_t_nrm if len(output_t_nrm) > 0 else output_t
		
		final_work_exp = {
			"title": final_title,
			"employer": final_org,
			"description": output_desc
		}
		work_exps.append(final_work_exp)
	POSITIONS = get_positions(work_exps)

	## EDUCATIONS
	EDUCATIONS = raw_data_item["educations"]
	
	## SKILLS
	SKILLS = raw_data_item["skills"]

	## EXPERIENCE SUMMARY
	EXP_SUMMARY = raw_data_item["experience_summary"]

	## CRITIQUE
	CRITIQUE = raw_data_item["critique"]

	data_item = {
		"resume_number": file_n,
		"name": NAME,
		"summary": SUMMARY,
		"profession": PROFESSION,
		"positions": POSITIONS,
		"exp_summary": EXP_SUMMARY,
		"educations": EDUCATIONS,
		"skills": SKILLS,
		"critique": CRITIQUE
	}
	
	return data_item
````

````python
def extraction_resume_prompt(resume):
    EXT_TEMPLATE = TEMPLATES["extraction_resume"]["prompt"]
    prompt = EXT_TEMPLATE.format(resume=resume)
    return prompt
````

````python
def process_response_azure_openai(response):
    res_json = json.loads(response.json())
    output_text = res_json["choices"][0]["message"]["content"]
    output = output_text.replace("```json", "").replace("```", "").strip()
    n_output_tokens = res_json["usage"]["completion_tokens"]
    try:
        output_json = json.loads(output)
    except:
        try:
            output = output.replace("': ", '": ').replace(", '", ', "').replace("{'", '{"').replace("'}", '"}')
            output_json = json_repair.repair_json(output, return_objects=True)
        except JSONDecodeError as e:
            output_json = None
    return output_json, output, n_output_tokens
````

````python
def invoke_azure_openai_with_text(primer: str, prompt: dict, temperature=0.1, max_tokens=16384, max_retries=5):
    file_n = prompt["file_n"]
    user_prompt = prompt["prompt"]
    messages = [
                {
                    "role": "user",
                    "content": user_prompt
                }
            ]
    completion = azure_openai_client.chat.completions.create(
        model="gpt-4o",
        messages=messages,
        temperature=temperature,
        max_tokens=max_tokens
    )
    output_json, output, n_output_tokens = process_response_azure_openai(completion)
    
    final_res = {
        "output_json": output_json,
        "file_n": file_n
    }
    return final_res
````

````python
PRIMER = TEMPLATES["extraction_resume"]["primer"]

def execute_azure_openai_with_text_in_thread_pool(prompts, temperature=0.1, max_tokens=16384, max_workers=3):
    """
    Executes the Claude model with multiple user prompts concurrently using a thread pool.

    Args:
        prompts (list): A list with records of prompts
        temperature (float, optional): The temperature parameter for the model. Defaults to 0.1.
        max_tokens (int, optional): The maximum number of tokens to generate. Defaults to 8192.

    Returns:
        list: A list of tuples containing the JSON response, the generated text, and the CSV name.
    """
    global PRIMER
    try:
        with tqdm(total=len(prompts)) as pbar:
            results = []
            with ThreadPoolExecutor(max_workers=max_workers) as executor:
                futures = [executor.submit(invoke_azure_openai_with_text, PRIMER, prompt, temperature, max_tokens) for prompt in prompts]
                # results = [future.result() for future in futures]
                for future in as_completed(futures):
                    pbar.update(1)
                    results.append(future.result())
        return results
    except Exception as e:
        print(f"Error executing thread pool. Error: {e}")
        return []
````
